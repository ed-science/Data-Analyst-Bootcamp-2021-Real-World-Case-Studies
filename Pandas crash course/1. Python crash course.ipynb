# Classes

```python
class Book(object):
  """ Class object for book """

  def __init__(self, title, author):
    """ Initializes a new book """
    self.title = title
    self.author = author
  def __str__(self):
    return self.title
  def change_book_name(self, new_name):
    self.title = new_name
```

```python
mybook = Book(title = "This is a book", author = "JN")
```

```python
print(mybook)
```

```python
mybook.change_book_name(new_name = 'This is a new name')
```

```python
print(mybook.title)
```
