# Pandas Feature Engineer
- Lambda function
- Apply function

```python
# Feature Engineering is the technique of extracting  more usefull features from existing data

# The wiki difination is "Feature engineering is the process of using domain knowledge of the data"
# to create feature that make machine learning algorithms works
# Let's re-load our Titantic dataset

import pandas as pd

file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv"
df = pd.read_csv(file_name)

df.head()
```

- sibsp: # of siblings/spouse abroad
- parch: number of parents/child aboard

```python
# Lambda expressions to create new features

# Let's define a  function that we will use to get the family size
def get_family_size(sibsp, parch):
  return sibsp + parch

df['family size'] = df[['sibsp', 'parch']].apply(lambda x: get_family_size(x['sibsp'], x['parch']), axis =1)
```
