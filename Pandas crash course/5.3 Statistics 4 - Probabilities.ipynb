# Probabilities


```python
from numpy.random import seed, randint

seed(1)

def coin_toss(flips):
  head = 0
  tail = 0
  for i in range(0,flips):
    if randint(0,2) == 1:
      head += 1
    else:
      tail += 1
  return head/flips

coin_toss(10000)
```

# Let's store the result of coin toss simulations

```python
probabilities = []
for i in range(1,1000):
  result = (i,coin_toss(i))
  probabilities.append(result)
```

```python
# Convert our data to a dataframe
import pandas as pd

df = pd.DataFrame(probabilities)
df.columns = ['Number of Coin tosses', 'Probability']
df.head(15)
```

```python
# Now let's create our plot

import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(x = 'Number of Coin tosses', y= 'Probability', ci=None, size=1, kind = 'line', data =df)
plt.axhline(0.5, color='red', label= 0.5)
plt.legend()
plt.show()
```
