# Let's understand concatenating

```python
# Let's create two dummy data frames to understand
import pandas as pd

df1 = pd.DataFrame({'item':['A','B','C','D'],
                    'value':[1,2,3,5]})
df2 = pd.DataFrame({'item':['E','F','G','H'],
                    'value':[5,6,7,8]})

print(df1)
print(df2)
```

```python
# Let's stack them vertically by using pd.concat

pd.concat([df1,df2]).reset_index()
```

```python
# What if they had different column names

df1 = pd.DataFrame({'item': ['A', 'B', 'C', 'D'],
                     'value': [1, 2, 3, 5]})

df2 = pd.DataFrame({'item': ['E', 'F', 'G', 'H'],
                    'quanity': [2, 2, 1, 5]})

pd.concat([df1, df2]).reset_index()
```

```python
# We can use asix = 1 to stack horizontally

pd.concat([df1,df2], axis =1)
```

# Similarly we can use Append to add multiple dataFrames on to another

```python
df1 = pd.DataFrame({'item': ['A', 'B', 'C', 'D'],
                     'value': [1, 2, 3, 5]})

df2 = pd.DataFrame({'item': ['D', 'F', 'G', 'H'],
                    'quanity': [2, 2, 1, 5]})

df3 = pd.DataFrame({'item': ['I', 'J', 'K', 'L'],
                    'quanity': [3, 4, 7, 25]})

df1.append([df2, df3]).reset_index()
```

# Pandas merge

Pandas has full-feature,high performance in-money join operations idiomatically very similar to relational database like SQL:
- **one-to-one join**: for example when joining two DataFrame object on their indexes (which must contain  unique values).
- many-to-one join: for example when joining an index(unique) to one or more columns in a different DataFrame.
- many-to-many joins: joining columns on columns.

![alt text](https://i.stack.imgur.com/hMKKt.jpg)

```python
dfA = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
dfB = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
dfA
```

```python
dfB
```

```python
# Merging on unique keys
# by defaut we're doing an inner join

pd.merge(dfA,dfB, on='key')
```

```python
# Understand the different between left and right joins

dfA = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

dfB = pd.DataFrame({'key': ['K4', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
pd.merge(dfA, dfB, how='left', on= 'key')
```

```python
pd.merge(dfA, dfB, how='right', on= 'key')
```

```python
pd.merge(dfA, dfB, how='inner', on= 'key')
```

# Joining

```python
left = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                    index=['K0','K1','K2'])
left
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                       'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
right
```

```python
left.join(right)
```

```python
left.join(right, how= 'outer')
```
